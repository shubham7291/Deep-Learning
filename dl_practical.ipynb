```python
import numpy as np
def f(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10):
    return ((x1^x2) & (x3 |x4)) ^((x5&x6) |((x7^x8) &(x9^x10)))
def get_Dataset():
    m=1024
    X=np.zeros((m,10))
    Y=np.zeros((m,1))
    for i in range(m):
        bit_Str=np.binary_repr(i,width=10)
        bits=np.array([int(b) for b in bit_Str])
        X[i,:]=bits
        Y[i,0]=f(*bits)
    return X,Y
x_data,y_data=get_Dataset()
```
